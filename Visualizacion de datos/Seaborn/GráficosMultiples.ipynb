```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv')
data.head()
```

### Gráficos múltiples.

```python
g = sns.FacetGrid(data=data, col='sex')
g.map(sns.histplot, 'age')
```

```python
g = sns.FacetGrid(data=data, col='pclass', hue='sex')
g.map(sns.scatterplot, 'age', 'fare')
```

```python
g = sns.FacetGrid(data=data, col='pclass', hue='sex')
g.map(sns.scatterplot, 'age', 'fare', color='green', s=60) 

for ax in g.axes.flat:
    ax.axhline(y=300)
```

# PairGrid

```python
#Genera todas las combinaciones posibles entre unas variables dadas.
g = sns.PairGrid(data=data, vars=['pclass','age','fare'], hue='sex')
#En la diagonal configuramos histogramas y puntos en las no diagonales.
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
#g.map_lower (Para los puestos abajo)
#g.map_upper(Para los puestos arriba)
```
