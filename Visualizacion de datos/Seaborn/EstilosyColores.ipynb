```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv')
data.head()
```

Manejar estilos y colores en Seaborn.

```python
sns.boxplot(x='pclass', y='age', data=data)
```

```python
sns.set_style('dark')
sns.boxplot(x='pclass', y='age', data=data)
#Con despine modificamos los ejes.
sns.despine(left=True, offset=10, trim=True)
```

```python
sns.set_theme()
```

```python
#Cambiamos el contexto.
sns.set_context('talk')
sns.boxplot(x='pclass', y='age', data=data)
```

```python
#Podemos modificar parámetros de los contextos.
sns.set_context('notebook', font_scale=1.4, rc={"lines.linewidth":3.5})
sns.boxplot(x='pclass', y='age', data=data)
```

```python
#Para cambiar la paleta de colores.
sns.set_context('notebook')
sns.boxplot(x='pclass', y='age', data=data, palette=sns.color_palette("husl",8))
```

```python
#Podemos cambiar los colores de un mapa de calor con cmap.
sns.heatmap(data.corr(), cmap=sns.color_palette('Accent', as_cmap = True))

```
