```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/mpg.csv')
data.head()
```

Grafico relacional de puntos.

```python
data['origin'].value_counts()
```

```python
#Height aumenta el tamaño. Data agrega el dataset. Hue divide los graficos por el parametro del dataset.
#Style divide las variables en cruces, puntos... Size cambia el tamaño del punto en funcion de la cantidad.
sns.relplot(x = 'weight', y = 'horsepower', data = data, height=5, hue='origin', style='cylinders', size='cylinders')

```

```python
#Si utilizamos kind=line hace un grafico de lineas, que tiene los mismos parametros que el anterior.
sns.relplot(x = 'weight', y = 'horsepower', data = data, kind='line', hue='origin')
```
