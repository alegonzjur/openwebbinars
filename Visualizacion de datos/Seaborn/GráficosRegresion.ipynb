```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')
data.head()
```

Gráfico de regresión (regplot)

```python
fig , ax = plt.subplots(figsize=(10,5))
#El gráfico de regresion intenta sacar una recta de un gráfico de puntos. La sombra es un intervalo de confianza.
#Con line_kws podemos cambiar algunos parámetros como el color.
sns.regplot(x='total_bill', y='tip', data=data, ax=ax, line_kws={'color':'red'})
```

```python
#Otra forma de crear rectas de regresion. Order genera curvas en lugar de rectas.
sns.lmplot(x='tip', y='total_bill', data=data, hue='smoker', order=2)
```

```python
#Con col podemos separar los gráficos.
sns.lmplot(x='tip', y='total_bill', data=data, hue='smoker', col='sex')
```

```python
#Combina las distribuciones con una recta de regresion.
sns.jointplot(x='tip', y='total_bill', data=data, kind='reg')
```
