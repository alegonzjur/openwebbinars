```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv')
data.head()
```

# Catplots

```python
#Swarn evita que los valores se superpongan. Hue separa los valores por sexo.
sns.catplot(x='pclass', y='age', data=data, kind='swarm', hue = 'sex')
```

# Boxplots

```python
fig, ax = plt.subplots(figsize=(10,5))
#Introducimos el eje creado con matplotlib como parámetro. 
#El diagrama de cajas indica los 3 cuartiles junto con los valores anomalos (que están fuera).
sns.boxplot(x='pclass',y='age',data=data, ax=ax)
```

```python
#Se puede hacer mediante catplot indicando  kind = box. Podemos cambiar la orientacion con orient. Recordar intercambiar las variables.
sns.catplot(x='age', y='pclass', data=data, kind='box',height=6, hue='sex',orient='h')
```

```python
#Grafico de violin.
sns.catplot(x='pclass', y='age', data=data, kind='violin',height=6, hue='sex')
```

```python
#Con bar podemos crear graficos de barras con catplot, y con el estimator podemos realizar un conteo.
sns.catplot(x='pclass',y='survived', data=data, kind='bar', estimator=np.count_nonzero)
```
