```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv')
data.head()
```

## Gr√°ficos para distribuciones.

Distplot.

```python
fig, ax = plt.subplots(figsize=(10,5))
#Para generar un distplot utilizamos lo siguiente. Se le puede a√±adir un subplot con ax. Norm_hist normaliza los valores.
#Bins expone el histograma en los valores introducidos.
sns.distplot(a = data['age'],bins=[5,10,30,50,90], hist = True, kde = False, ax=ax, norm_hist=True)
ax.set_title("Distribucion edad")
```

```python
data['pclass'].value_counts()
```

```python
sns.distplot(a=data['pclass'], bins=3, hist=True, kde=False)
```

```python
#Element = step hace que no se solapen.
sns.displot(data, x='age', hue='sex', height=7, element='step')
```

```python
#Hace que se pongan una variable encima de otra.
sns.displot(data, x='age', hue='sex', height=7,multiple='stack')
```

```python
#Con dodge hace que se pongan una al lado de otra.
sns.displot(data, x='age', hue='sex', height=7, multiple='dodge')
```

```python
#Genera un gr√°fico por cada valor de la variable introducida en col. En este caso 2. 
sns.displot(data, x="age", col = 'sex', height=7, stat = 'density')
```

## KDE (Kernel Density Function)

```python
fig , ax = plt.subplots(figsize=(10,5))
sns.distplot(a = data['age'], bins=10, hist=False, kde=True, ax = ax)
```

```python
sns.displot(data, x='age', height=7, kind='kde', hue='sex', bw_adjust = .25)
```

## ECDF

```python
sns.displot(data, x='age', kind='ecdf', hue='sex')
#Se puede interpretar como que el 20%(0.2) son menores de 20 a√±os. Con hue lo dividimos por sexo.
```
