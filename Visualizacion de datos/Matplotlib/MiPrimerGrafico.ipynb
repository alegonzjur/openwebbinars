### Mi primer gráfico.

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import matplotlib.style as mplstyle

import warnings 
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
x = np.arange(5)
y = np.array([2.0,2.5,1.5,3.0,4.5])
```

Montar un gráfico de línea a partir de 2 arrays.

```python
plt.plot(x,y)
```

Grafico de puntos.

```python
plt.plot(x,y,'o') #Si ponemos xr será de cruces.
```

```python
z = y + 1
z
```

```python
w = np.log(y)
```

```python
plt.plot(x,y,'*', label='y')
plt.plot(x,z,'g-', label='z')
plt.plot(x,w,'--', label = 'w')
plt.legend() #Añade una leyenda
plt.title('Mi primer grafico') #Añade un titulo
plt.xlabel('Eje x') #Añade un nombre al eje x.
plt.ylabel('Eje y') #Añade un nombre al eje y.
```

```python
mplstyle.use('ggplot')

plt.plot(x,y,'*', label='y')
plt.plot(x,z,'g-', label='z')
plt.plot(x,w,'--', label = 'w')
plt.legend() #Añade una leyenda
plt.title('Mi primer grafico') #Añade un titulo
plt.xlabel('Eje x') #Añade un nombre al eje x.
plt.ylabel('Eje y') #Añade un nombre al eje y.
```
