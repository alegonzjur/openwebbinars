Consultamos datos con ISIN

```python

dataset[dataset['Age'].isin([41,42])]
```

```python
dataset['Age'].where(dataset['Age'] == 41)
#Todo lo que no sea la edad configurada lo pone a Nan
```

```python
dataset['Age'].where(dataset['Age'] == 41, -1)
#Podemos cambiar el Nan por otro valor
```
