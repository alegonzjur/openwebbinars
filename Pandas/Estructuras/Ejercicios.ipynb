### Ejercicio 4.
- Convierte la columna 'date' en un tipo fecha.
- Genera 3 nuevas columnas fruto de extraer el año, mes y dia.
- ¿Cual es el precio medio por mes?
- Crea una nueva columna fruto de sumar 20 días a la columna fecha.

```python
data = pd.read_csv(data_path / 'house.csv')
data.head(2)
```

```python
data['date'] = pd.to_datetime(data['date'], yearfirst=True)
data.head()
```

```python
data['date']
```

```python
data['year'] = data['date'].dt.year
data['month'] = data['date'].dt.month
data['day'] = data['date'].dt.day
```

```python
data.head()
```

```python
data.set_index(data['date']).resample('1M')['price'].mean()
```

```python
td = pd.Timedelta('20 days')
td
```

```python
data['date2'] = data['date'] + td 
data[['date','date2']].head(2)
```
